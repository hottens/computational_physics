## Solving equations of motion of a constrained mechanical systems

The vibrations or oscillations of mechanical systems constitute one of the most important fields of study in physics. Virtually every system possesses the capability for vibrations, and most systems can vibrate in a large variety of ways. A continuous system has infinite degrees of freedom so it is always useful to model these systems with simple mechanical models with lower degrees of freedom. In this homework you will solve the equations of motion of a two-degrees-of-freedom mechanical system containing mass, spring, and damper using Runge-Kutta method.

# Q1

````python
def F(t):
    return 0

def f(y, t, k1, k2, c, m1, m2):
    f1 = y[2]
    f2 = y[3]
    f3 = (-k1*y[0] + k2*(y[1]-y[0]) + c*(y[3]-y[2]))/m1
    f4 = (F(t) - k2*(y[1]-y[0]) - c*(y[3]-y[2]))/m2
    return np.array([f1, f2, f3, f4])


def rk4(h, y, f, k1, k2, c, m1, m2, t_max):
    for t in np.arange(h,t_max,h):
        
        k_1 = h*f(y[-1],       t, k1, k2, c, m1, m2)
        k_2 = h*f(y[-1]+k_1/2, t+h/2, k1, k2, c, m1, m2)
        k_3 = h*f(y[-1]+k_2/2, t+h/2, k1, k2, c, m1, m2)
        k_4 = h*f(y[-1]+k_2,   t+h, k1, k2, c, m1, m2)

        y.append(y[-1]+(k_1+2*k_2+2*k_3+k_4)/6)


    return np.array(y), np.arange(0,t_max,h)
````

$$
\begin{aligned}
    det(K-M\omega^2) = 0 &= (k_1+k_2 - m_1 \cdot \omega^2) \cdot (k_2 - m_2 \cdot \omega^2) - k_2^2 \\
  &= \omega^4 \cdot m_1m_2 - \omega^2(m_2k_2 + m_2k_1 + m_1k_2) + k_2k_1 \\
  &= 1\omega^4 + 0 \omega^3 -15\omega^2 + 0\omega +25
\end{aligned}
$$

````python
m1, m2 = 1, 1
k1, k2 = 5, 5

M = np.array([[m1, 0], [0, m2]])
K = np.array([[k1+k2, -k2], [-k2, k2]])

# det(K-Mw^2) = (k1+k2 - m1w^2) * (k2-m2w^2) - k2^2 = 0
#           0 = w^4 * m1*m2 - w^2*(m2k2 + m2k1 + m1k2) + k2k1
#             = 1 w^4 + 0 w^3 -15 w^2 + 0 w + 25 

w = np.roots([1, 0, -15, 0, 25])
nfs = w[w > 0]/(2*np.pi)
print("The natural frequencies are {} Hz".format(nfs))
````

# Q8

````python
def F(t):
    return 1.5 * np.sin(t)


init_conditions = [(5, 5, 2, 1, 1), (10, 5, 2, 1, 1), (5, 10, 2, 1, 1), (5, 5, 4, 1, 1), (5, 5, 2, 1, 3), (5, 5, 2, 3, 1)]
condition_names = ["base", "increased k1", "increased k2", "increased c", "increased m2", "increased m1"]
stable_times = []

for name, conditions in zip(condition_names, init_conditions):
    k1, k2, c, m1, m2 = conditions
    y0= [np.array([5, 10, 0, 0])]
    data, t = rk4(0.1, y0, f, k1, k2, c, m1, m2, 700)

    v1 = data[:,0]
    v2 = data[:,1]
    x1 = data[:,2]
    x2 = data[:,3]
    
    # Compare positive peaks:
    # peaks are where the velocity is roughly 0
    mask = (np.round(v1, decimals=1)==0) & (x1>0)
    
    # Allow a small error
    e = 5e-3
    
    # We can say that stability is reached the first time the next peak is the same AND the 10th peak after too
    t_stable = t[mask][
        (abs(x1[mask] - np.roll(x1[mask], -1)) < e) &
        (abs(x1[mask] - np.roll(x1[mask], -10)) < e)
    ][0]
    
    t_idx = np.where(t==t_stable)[0][0]
    
    stable_times.append(t_stable)
    
    figX = plt.figure()
    plt.plot(t[0:min(2*t_idx, len(t))],x1[0:min(2*t_idx, len(t))])
    #plt.plot(t,x2)
    plt.xlabel("t")
    plt.ylabel("x")
    plt.axvline(t_stable, ls="--", color="black")
    plt.title(name)
    #plt.legend(["x1", "x2", "t_stable"])
    plt.legend(["x1", "t_stable"])
    plt.show()
    print("Stable at t = {}".format(t_stable))
    if len(stable_times) > 1:
        
        print("Which is {} seconds {} than the base measurement.".format(abs(stable_times[0]-t_stable), "longer" if t_stable > stable_times[0] else "shorter"))
````

Above, we see what effect modifying individual parameters has on the stabilizing time. We compare different setups against the base setup.

When we increased only spring constant $k_1$, we see that stabilization is reached earlier.
Presumably, the higher this spring constant, the stiffer the spring and the less mass1 is able to move, causing the amplitude to decrease.

Increasing spring constant $k_2$, though, the stable state is reached later than the base measurement.
By increasing this spring constant, which is connected to mass 1 and 2, it is, apparently, harder for the system to get in harmony. 

Increasing damping parameter $c$, causes the system to stabalize earlier, since higher damping results in less energy transfer between the two masses.

Increasing mass $m_2$ results in a longer stabilization time. It takes a little bit longer for mass 1 and 2 to move in syncrony when mass 2 gets heavier.

Increasing mass $m_1$ by the same scale, though, stabilization is reached way later. The interaction between mass 1 and 2 depend more on mass 1 now, because it's heavier. Since two springs are attached to mass 1, the system needs to go through multiple transitions of amplitudes before a stabilization is achieved.
